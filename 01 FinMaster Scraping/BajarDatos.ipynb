```python
remax_detalle= db.RemaxDetalle
```

```python
#marca los unicos y borra los duplicados del detalle
bikes_list = list(remax_collection.find({'deta': {'$ne': 1 }},{'id'}))
for num in bikes_list:
    cantidad =1
    print("ID "+ num['id'])
    for prop in remax_detalle.find({'id':'ID: '+num['id']},{'_id', 'id'}):
        if cantidad==1:
            print("Unico "+ num['id'])
            remax_collection.update_many({'id' : num['id']},{'$set': {'deta' : int(1)}}) 
            remax_detalle.update_many({'_id' : prop['_id']},{'$set': {'id1' : str(num['id'])}}) 
            cantidad+=1
        else:
            print("   del: "+prop['id'])
            remax_detalle.delete_many({'_id' : prop['_id']}) 
```

```python
#marca los unicos y borra los duplicados
bikes_list = list(remax_collection.find({ 'unic': {'$ne': 1 } },{'id'}))
for num in bikes_list:
    cantidad =1
    for prop in remax_collection.find({'id':num['id']},{'_id', 'id'}):
        if cantidad==1:
            print("Unico "+ num['id'])
            remax_collection.update_many({'_id' : prop['_id']},{'$set': {'unic' : int(1)}}) 
            cantidad+=1
        else:
            print("   del: "+prop['id'])
            remax_collection.delete_many({'_id' : prop['_id']}) 
        
```

```python
unicos = list(remax_collection.find({ 'unic':1 },{'id'}))
for num in unicos:
    print("Unico "+ num['id'])
    for prop in remax_collection.find({'_id' : {'$ne': num['_id'] } , 'id':num['id']},{'_id', 'id'}):
        print("   del: "+prop['id'])
        remax_collection.delete_many({'_id' : prop['_id']}) 
```

```python
bikes_list = list(remax_collection.distinct('tipoinv'))   
for num in bikes_list:  
    if num=="Solicitar entrevista":
        remax_collection.update_many({'prize' : num},{'$set': {'precio' : float(0)}}) 
    else:
        remax_collection.update_many({'meters' : num},{'$set': {'mts' : float(num.replace(',', ''))}}) 
```
